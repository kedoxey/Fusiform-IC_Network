```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cell_numbers = np.arange(15)
weights = [1, 2, 3, 3, 3, 2, 1, 1, 1, 2, 3, 3, 3, 2, 1]
```

```python
fig, axs = plt.subplots(1, 1, figsize=(8,4))
axs.plot(cell_numbers, weights, '-o')
axs.set_yticks([1, 2, 3])
axs.set_ylabel('Relative Inhibition Strength')
axs.set_xticks([0, 3, 6, 7, 8, 11, 14])
axs.set_xticklabels(['n - 7', 'n - 4', 'n - 1', 'n', 'n + 1', 'n + 4', 'n + 7'])
axs.set_xlabel('Cell (id)')

fig.tight_layout()
fig.savefig('relative_inh_strength.png', dpi=300)
```
